```python
import pandas as pd

BASE = "/content/drive/MyDrive/submission/data"

products = pd.read_csv(f"{BASE}/products_dataset.csv")
translation = pd.read_csv(f"{BASE}/product_category_name_translation.csv")

print("products:", products.shape)
print("translation:", translation.shape)

category_map = dict(zip(
    translation["product_category_name"],
    translation["product_category_name_english"]
))

products["product_category_name"] = products["product_category_name"].map(category_map)

# isi missing
products["product_category_name"] = products["product_category_name"].fillna("others")

print("\nmissing category after map:", products["product_category_name"].isna().sum())
print(products["product_category_name"].value_counts().head(10))

# export CLEAN
out = f"{BASE}/products_dataset_clean.csv"
products.to_csv(out, index=False)
print("\nsaved:", out)
```
