## Data Preparation / Preprocessing

```python
# Copy dataset agar raw tetap aman
df_clean = df.copy()

# A. BASIC CLEANING

# 1. Hapus duplicate
before_rows = len(df_clean)
df_clean = df_clean.drop_duplicates()
after_rows = len(df_clean)
print(f"Duplicate rows removed: {before_rows - after_rows}")

# 2. Hapus kolom ID unik jika ada
if 'EmployeeId' in df_clean.columns:
    df_clean = df_clean.drop(columns=['EmployeeId'])
    print("Column 'EmployeeId' removed.")

# 3. Hapus kolom konstan (nilai hanya 1 jenis)
constant_cols = [col for col in df_clean.columns if df_clean[col].nunique(dropna=False) <= 1]
if constant_cols:
    df_clean = df_clean.drop(columns=constant_cols)
    print(f"Constant columns removed: {constant_cols}")
else:
    print("No constant columns found.")

# B. HANDLE MISSING VALUES

# Kolom numerik dan kategorikal
num_cols = df_clean.select_dtypes(include=[np.number]).columns
cat_cols = df_clean.select_dtypes(include=['object']).columns

# Isi missing numerik dengan median
for col in num_cols:
    if df_clean[col].isnull().sum() > 0:
        df_clean[col] = df_clean[col].fillna(df_clean[col].median())

# Isi missing kategorikal dengan modus
for col in cat_cols:
    if df_clean[col].isnull().sum() > 0:
        mode_val = df_clean[col].mode()
        if len(mode_val) > 0:
            df_clean[col] = df_clean[col].fillna(mode_val[0])

print("Missing values handled.")

# C. SIMPLE TRANSFORM FOR TABLEAU

# 1. Standardize Attrition => Yes / No
if 'Attrition' in df_clean.columns:
    if pd.api.types.is_numeric_dtype(df_clean['Attrition']):
        df_clean['Attrition'] = df_clean['Attrition'].apply(lambda x: 'Yes' if x == 1 else 'No')
    else:
        df_clean['Attrition'] = df_clean['Attrition'].astype(str).str.strip().str.title()
    print("Column 'Attrition' standardized to Yes/No.")

# 2. Standardize OverTime
if 'OverTime' in df_clean.columns:
    df_clean['OverTime'] = df_clean['OverTime'].astype(str).str.strip().str.title()
    print("Column 'OverTime' standardized.")

# 3. Age Group
if 'Age' in df_clean.columns:
    df_clean['AgeGroup'] = pd.cut(
        df_clean['Age'],
        bins=[0, 25, 35, 45, 55, 100],
        labels=['<=25', '26-35', '36-45', '46-55', '55+'],
        include_lowest=True
    )
    print("Column 'AgeGroup' created.")

# 4. Years At Company Group
if 'YearsAtCompany' in df_clean.columns:
    df_clean['YearsAtCompanyGroup'] = pd.cut(
        df_clean['YearsAtCompany'],
        bins=[-1, 2, 5, 10, 20, 100],
        labels=['0-2 Years', '3-5 Years', '6-10 Years', '11-20 Years', '20+ Years'],
        include_lowest=True
    )
    print("Column 'YearsAtCompanyGroup' created.")

# 5. Monthly Income Group (aman dari error bin duplicate)
if 'MonthlyIncome' in df_clean.columns:
    income_max = df_clean['MonthlyIncome'].max()

    if income_max <= 3000:
        bins = [0, income_max + 1]
        labels = ['<=3K']
    elif income_max <= 6000:
        bins = [0, 3000, income_max + 1]
        labels = ['<=3K', '3K+']
    elif income_max <= 10000:
        bins = [0, 3000, 6000, income_max + 1]
        labels = ['<=3K', '3K-6K', '6K+']
    elif income_max <= 15000:
        bins = [0, 3000, 6000, 10000, income_max + 1]
        labels = ['<=3K', '3K-6K', '6K-10K', '10K+']
    else:
        bins = [0, 3000, 6000, 10000, 15000, income_max + 1]
        labels = ['<=3K', '3K-6K', '6K-10K', '10K-15K', '15K+']

    df_clean['MonthlyIncomeGroup'] = pd.cut(
        df_clean['MonthlyIncome'],
        bins=bins,
        labels=labels,
        include_lowest=True
    )
    print("Column 'MonthlyIncomeGroup' created.")

# D. FINAL CHECK

print("\nPreview cleaned dataset:")
display(df_clean.head())

print(f"\nFinal dataset shape: {df_clean.shape}")

print("\nRemaining missing values:")
display(df_clean.isnull().sum())

# E. EXPORT FOR TABLEAU

# CSV backup
csv_file = 'fact_tableau.csv'
df_clean.to_csv(csv_file, index=False)

# TSV recommended untuk Tableau, CSV delimiter bermasalah
tsv_file = 'fact_tableau.tsv'
df_clean.to_csv(tsv_file, index=False, sep='\t')

print(f"\nFiles created successfully:")
print(f"- {csv_file} (backup)")
print(f"- {tsv_file} (recommended if Tableau has delimiter issue)")

# Download file Tableau Friendly wkwk
files.download(tsv_file)
```
